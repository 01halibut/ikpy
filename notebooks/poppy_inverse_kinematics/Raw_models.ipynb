First, import dependencies

```python
%matplotlib inline
import poppy_inverse_kinematics.robot_utils
import poppy_inverse_kinematics.model
import poppy_inverse_kinematics.model_config
```

Then, create a config object, from a URDF file

```python
params = poppy_inverse_kinematics.model_config.from_urdf_file(urdf_file="../resources/poppy_ergo.URDF", base_link="base_link", last_link_vector=[0.05, 0, 0])
```

Finally, create your model, from your config object.

```python
robot = poppy_inverse_kinematics.model.Model(params, computation_method="hybrid", simplify=False)
```

Define the target of your robot

```python
target = [-0.1, 0.1, 0.1]
robot.target = target
```

Use Inverse Kinematics to reach your target

```python
robot.goto_target()
```

Display the result

```python
robot.plot_model()
print("Robot position : {} / Original target : {}".format(robot.forward_kinematic(), robot.target))
```
